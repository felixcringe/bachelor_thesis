```python
#! /usr/bin/env python
# -*- coding: utf8 -*-

import NowcastingSuite

import os


def simple_usr_prompt(accepted_answers=['yes', 'no', 'y', 'n'], prompt_str='>', show_answers=True):
    usr_inp = ''
    while usr_inp == '':
#         usr_inp = raw_input(prompt_str)
        usr_inp = input(prompt_str)
        if usr_inp.lower() not in map(lambda o: str(o).lower(), accepted_answers):
            print ("ERROR: Invalid input '%s'! Please try again!" % (usr_inp)),
            if show_answers:
                print (" Valid answers are:", " ".join(accepted_answers))
            else:
                print ("")
            usr_inp = ''
        else:
            return usr_inp

def check_folders():
    _data_dir = './data/'
    _plots_dir = './plots/'
    if not os.path.exists(_data_dir):
        raise ValueError("ERROR: Data folder ('%s') is not present here!" % (_data_dir))
    if not os.path.exists(_plots_dir):
        print ("INFO: Plot folder ('%s') not present, trying to make a new folder..." % (_plots_dir))
        os.mkdir(_plots_dir)
        if not os.path.exists(_plots_dir):
            raise OSError("ERROR: Couldn't make dir '%s'!" % (_plots_dir))


# if __name__ == "__main__":
print ("Nowcasting Example\n------------")
check_folders()
print ("Please choose an event:")
print ("A: 2016-6-24, radar Essen, 10UTC - 15UTC")
print ("B: 2016-6-24, radar Neuheilenbach, 09:15UTC - 12:00UTC")
print ("C: 2016-6-24, radar Türkheim, 15UTC - 16UTC")
print ("")
_stop_words = ['quit', 'exit', 'abort', 'stop']
choice = simple_usr_prompt(['A', 'B', 'C',]+_stop_words)
if choice in _stop_words:
    print ("quitting...")
    import sys
    sys.exit()
else:
    event_dt, chosen_radar, frames_dt_objs = NowcastingSuite.get_event_infos(choice)
    NowcastingSuite.run_offline_event(event_dt, chosen_radar, frames_dt_objs, _save_attcor_plots=True, _save_hafer_plots=True)
    #
#

```
